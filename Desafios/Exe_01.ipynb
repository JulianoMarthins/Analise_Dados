```python
import pandas as pd 

# Importando base de dados
funcionario_df = pd.read_csv("CadastroFuncionarios.csv", sep=';', decimal=',')
cliente_df = pd.read_csv("CadastroClientes.csv", sep=';')
servico_df = pd.read_excel("BaseServiçosPrestados.xlsx")

# Renomeando colunas para padronização
servico_df = servico_df.rename(columns={'Tempo Total de Contrato (Meses)': 'Tempo de Contrato'})
cliente_df = cliente_df.rename(columns={'Valor Contrato Mensal': 'Mensalidade'})


# Selecionando colunas úteis a nossa análise
funcionario_df = funcionario_df.drop(["Estado Civil"], axis=1)
servico_df = servico_df.drop(['Codigo do Servico'], axis=1)

# associando tabelas
cliente_df = cliente_df.merge(servico_df, on='ID Cliente')

print('             Serviços Prestados')
display(servico_df)
print('\n            Clientes da empresa')
display(cliente_df)
print('\n            Funcionarios')
display(funcionario_df)
```

```python
# 1. Valor total da folha salarial.

salario = funcionario_df['Salario Base'].sum()
impostos = funcionario_df['Impostos'].sum()
beneficios = funcionario_df['Beneficios'].sum()
vt = funcionario_df['VT'].sum()
vr = funcionario_df['VR'].sum()

salario_total = salario + impostos + beneficios + vt + vr


print(f'Salário Total: R$ {salario_total:,.2f}')
```

```python
# 2. Calcular o faturamento total da empresa
mensalidade = cliente_df['Mensalidade'].sum()
meses_contrato = cliente_df['Tempo de Contrato'].sum()



faturamento = (mensalidade * meses_contrato) - salario_total

print(f'Faturamento: R$ {faturamento:,.2f}')
```
